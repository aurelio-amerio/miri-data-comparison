```python
import numpy as np
import astropy.io.fits as fits
import matplotlib.pyplot as plt
```

```python
filename_ch1 = "jw01282-o032_t064_miri_ch1-short_x1d.fits"
filename_ch2 = "jw01282-o032_t064_miri_ch2-short_x1d.fits"

file_new_ch1 = fits.open("miri_data_comparison/data/new/"+filename_ch1)
file_old_ch1 = fits.open("miri_data_comparison/data/old/"+filename_ch1)

file_new_ch2 = fits.open("miri_data_comparison/data/new/"+filename_ch2)
file_old_ch2 = fits.open("miri_data_comparison/data/old/"+filename_ch2)
```

```python
file_new_ch1[0].header["DATE-BEG"], file_new_ch2[0].header["DATE-BEG"]
```

```python
file_old_ch1[0].header["SDP_VER"], file_old_ch1[0].header["CAL_VER"], file_new_ch1[0].header["SDP_VER"], file_new_ch1[0].header["CAL_VER"]
```

```python
file_old_ch1[0].header["VISITEND"], file_new_ch1[0].header["VISITEND"]
```

```python
data_old_ch1 = file_old_ch1[1].data
data_new_ch1 = file_new_ch1[1].data 

data_old_ch2 = file_old_ch2[1].data
data_new_ch2 = file_new_ch2[1].data
```

```python
wavelength_old_ch1 = data_old_ch1["WAVELENGTH"]
wavelength_new_ch1 = data_new_ch1["WAVELENGTH"]

flux_old_ch1 = data_old_ch1["FLUX"]
flux_new_ch1 = data_new_ch1["FLUX"]

flux_err_old_ch1 = data_old_ch1["FLUX_ERROR"]
flux_err_new_ch1 = data_new_ch1["FLUX_ERROR"]

bkd_old_ch1 = data_old_ch1["BACKGROUND"]
bkd_new_ch1 = data_new_ch1["BACKGROUND"]

bkd_err_old_ch1 = data_old_ch1["BKGD_ERROR"]
bkd_err_new_ch1 = data_new_ch1["BKGD_ERROR"]

wavelength_old_ch2 = data_old_ch2["WAVELENGTH"]
wavelength_new_ch2 = data_new_ch2["WAVELENGTH"]

flux_old_ch2 = data_old_ch2["FLUX"]
flux_new_ch2 = data_new_ch2["FLUX"]

flux_err_old_ch2 = data_old_ch2["FLUX_ERROR"]
flux_err_new_ch2 = data_new_ch2["FLUX_ERROR"]

bkd_old_ch2 = data_old_ch2["BACKGROUND"]
bkd_new_ch2 = data_new_ch2["BACKGROUND"]

bkd_err_old_ch2 = data_old_ch2["BKGD_ERROR"]
bkd_err_new_ch2 = data_new_ch2["BKGD_ERROR"]


```

```python
plt.plot(wavelength_old_ch1, flux_old_ch1, label="old")
plt.plot(wavelength_new_ch1, flux_new_ch1, label="new")

plt.legend()
plt.xlabel(r"Wavelength [$\mu m$]")
plt.ylabel("Flux (Jy)")
plt.title("Flux Comparison")
plt.ylim(0,0.08)
plt.show()
```

```python
plt.plot(wavelength_old_ch1, bkd_old_ch1, label="old ch1")
plt.plot(wavelength_new_ch1, bkd_new_ch1, label="new ch1")

plt.legend()
plt.xlabel(r"Wavelength [$\mu m$]")
plt.ylabel("Background (MJy/sr)")
plt.title("Background Comparison")

# plt.plot(wavelength_old_ch2, bkd_old_ch2, label="old ch2")
# plt.plot(wavelength_new_ch2, bkd_new_ch2, label="new ch2")

# plt.legend()
# plt.xlabel(r"Wavelength [$\mu m$]")
# plt.ylabel("Background (MJy/sr)")
# plt.title("Background Comparison")
plt.show()
```

```python
plt.plot(wavelength_old_ch1, bkd_old_ch1, label="old")
plt.plot(wavelength_new_ch1, bkd_new_ch1, label="new")

plt.legend()
plt.xlabel(r"Wavelength [$\mu m$]")
plt.ylabel("Background (MJy/sr)")
plt.title("Background Comparison")
plt.show()
```

```python
plt.plot(wavelength_old_ch1, bkd_err_old_ch1, label="old")
plt.plot(wavelength_new_ch1, bkd_err_new_ch1, label="new")

plt.legend()
plt.xlabel(r"Wavelength [$\mu m$]")
plt.ylabel("Background uncertainty (MJy/sr)")
plt.title("Background Comparison")
plt.show()
```
